```python
from dataclasses import dataclass
```

```python
#entity 
@dataclass
class model_training_config:
    root_dir: Path
    updated_model_path: Path
    trained_model_mod: Path
    img_size: int
    data_yaml: Path
    mode: str
    model: str
    epochs: int
    plots: bool
    task: str
```

```python
#configuration

class ConfigurationManager:
    def __init__(self,config=CONFIG_FILE_PATH,params=PARAMS_FILE_PATH):
        self.config=read_yaml(config)
        self.params=read_yaml(params)
        
        create_directories([self.config.artifacts_root])
        
    def get_model_traning_config(self)->model_training_config:
        config=self.config.prepare_base_model
        
        create_directories([config.root_dir])
        
        model_return=model_training_config(
                root_dir=config.root_dir,
                updated_model_path=config.updated_model_path, 
                trained_model_mod=config.trained_model_mod,
                img_size=self.params.img_size,
                data_yaml=self.params.data_yaml,
                mode=self.params.mode,
                model=self.params.model,
                epochs=self.params.epochs, 
                plots=self.params.plots, 
                task=self.params.task     
                    )
        
        return model_return
        
```

```python
# component
import ultralytics
import os
import urllib.request as request
from zipfile import ZipFile
import tensorflow as tf
from ultralytics import YOLO
from ultralytics.yolo.v8.detect.predict import DetectionPredictor
import cv2 
```

```python
class Base_model:
    def __init__(self,config:model_training_config):
        self.config=config
        
    def model_tarining(self):
        command = f"yolo task={self.config.task} mode={self.config.mode} model={self.config.model} data={self.config.data_yaml} epochs={self.config.epochs} imgsz={self.config.img_size} plots={self.config.plots} save={self.config.updated_model_path}"
        !{command}
        
```
